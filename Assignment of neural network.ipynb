```python
forestfire1=forestfire.apply(LabelEncoder().fit_transform)
```

```python
forestfire1=forestfire1.drop(['month','day'],axis=1)
```

```python
forestfire1
```

```python
forestfire2=forestfire1.iloc[:,[4,5,6,7,8,28]]
```

```python
forestfire2
```

```python
#creating train and test column by using useful columns only as i mentioned above

forestfire3=forestfire2.values
```

```python
x=forestfire3[:,0:5]
x
```

```python
y=forestfire3[:,5]
y
```

```python
!pip install seaborn
```

```python
import seaborn as sns
sns.pairplot(forestfire2,hue='size_category')
```

```python
#creating model
model=Sequential()
model.add(Dense(10,input_dim=5,kernel_initializer='uniform',activation='relu'))
model.add(Dense(5,kernel_initializer='uniform',activation='relu'))
model.add(Dense(1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
# compile model:- means to say we check weather created is working properly or not
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=model.fit(x,y,validation_split=0.33,epochs=100,batch_size=10)
```

```python
scores=model.evaluate(x,y)
print('%s: %.2f%%'  % (model.metrics_names[1],scores[1]*100))
```

```python
model.history.history.keys()

```

```python
print(history.history.keys())
```

```python
import matplotlib.pyplot as plt
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
